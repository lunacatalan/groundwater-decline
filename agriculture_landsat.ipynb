## This function is for calculating the SAVI and NDVI of the buffered area 

```python
def savi_mean(search):
    
    bands_of_interest = ["nir08", "red"]
    
    data = odc.stac.stac_load(
        [search], 
        bands=bands_of_interest
    ).isel(time=0)
    
    # choose the bands from the data
    red = data["red"].astype("float")
    nir = data["nir08"].astype("float")
    
    # calculate the savi
    savi = ((nir - red) / (nir + red + 0.5))*1.5

    # calculate the ndvi
    ndvi = (nir - red) / (nir + red)

    # clip rasters
    savi_buffer = savi.rio.clip(radius.geometry)
    savi_quantile = savi_buffer.quantile([0, 0.25, 0.5, 0.75, 1])
    savi_sum = savi_buffer.sum() # keep it as an array
    savi_mean = savi_buffer.mean() # keep it as an array
    savi_df = pd.DataFrame(np.hstack((savi_quantile,  # stack the arrays
                                      savi_sum,
                                      savi_mean)), 
                              columns = [f"{search}_savi"])
    # clip rasters to aoi
    ndvi_buffer = ndvi.rio.clip(radius.geometry)
    ndvi_quantile = ndvi_buffer.quantile([0, 0.25, 0.5, 0.75, 1])
    ndvi_sum = ndvi_aquifer.sum()
    ndvi_mean = ndvi_aquifer.mean()
    ndvi_df = pd.DataFrame(np.hstack((ndvi_quantile, ndvi_sum, ndvi_mean)),
                           columns = [f"{search}_ndvi"])
    
    datasets = [savi_df, ndvi_df] # make list of dataframes
    buffer_df = pd.concat(datasets, axis = 1)

    return(buffer_df)
```

## This function is for calculating the SAVI and NDVI of the aquifer area

```python
def savi_mean_aquifer(search):
    
    bands_of_interest = ["nir08", "red"]
    
    data = odc.stac.stac_load(
        [search], 
        bands=bands_of_interest
    ).isel(time=0)
    
    # choose the bands from the data
    red = data["red"].astype("float")
    nir = data["nir08"].astype("float")
    
    # calculate the ndvi
    savi = ((nir - red) / (nir + red + 0.5))*1.5

    # calculate the ndvi
    ndvi = (nir - red) / (nir + red)

    # clip rasters
    savi_aquifer = savi.rio.clip(qazvin_aquifer.geometry)
    savi_quantile_aquifer = savi_aquifer.quantile([0, 0.25, 0.5, 0.75, 1])
    savi_sum = savi_aquifer.sum() # keep it as an array
    savi_mean = savi_aquifer.mean() # keep it as an array
    savi_aq_df = pd.DataFrame(np.hstack((savi_quantile_aquifer,  # stack the arrays
                                         savi_sum,
                                         savi_mean)), 
                              columns = [f"{search}_savi"])
    # clip rasters to aoi
    ndvi_aquifer = ndvi.rio.clip(qazvin_aquifer.geometry)
    ndvi_quantile = ndvi_aquifer.quantile([0, 0.25, 0.5, 0.75, 1])
    ndvi_sum = ndvi_aquifer.sum()
    ndvi_mean = ndvi_aquifer.mean()
    ndvi_df = pd.DataFrame(np.hstack((ndvi_quantile, ndvi_sum, ndvi_mean)),
                           columns = [f"{search}_ndvi"])
    
    datasets = [savi_aq_df, ndvi_df] # make list of dataframes
    aquifer_df = pd.concat(datasets, axis = 1)
    
    return(aquifer_df)
```
